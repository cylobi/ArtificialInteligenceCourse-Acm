```python
import matplotlib.pyplot as plt
import numpy as np
X = np.sort(np.random.rand((100)))
sin_sample = np.sin(1.5*np.pi * X)*5
gaussian_noise = np.sort(np.random.normal(1, 0.5, 100))
x = gaussian_noise
y = x + sin_sample 
plt.scatter(x, y, s = 10)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

degrees = [1, 3, 7, 11 ,16, 20]
colors= ["r","g", "y" ,"k" ,"m" , "c"]
labels = [str(x) for x in degrees] 

for i in range(6):
    plt.figure()

    linear_regression = LinearRegression()
    polynomial_features = PolynomialFeatures(degree=degrees[i],include_bias=False)

    model = Pipeline([("polynomial_features", polynomial_features),("linear_regression", linear_regression)])
    model.fit(x.reshape(-1, 1), y)

    y_pred=model.predict(x.reshape(-1, 1))
    x_test = np.linspace(-1, 2, 100)
    y_test_pred = model.predict(x_test.reshape(-1, 1))
    y_test_real = np.sin(np.pi * x_test)*5
    
    mse = mean_squared_error(y, y_pred)
    print('mse for degree' + labels[i] + '= ' + str(mse))

    plt.scatter(x, y,color='b', label="sample")
    plt.plot(x_test, y_test_real,label='sin plot')
    plt.plot(x_test, y_test_pred,linewidth = 2 ,color=colors[i], label=labels[i])

    plt.xlim((-1, 2))
    plt.ylim((-5, 8))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.title('Degree '+ labels[i])
    
    
```
